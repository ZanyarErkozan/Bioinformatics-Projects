Your semester project involves *forensic DNA kinship testing*. With Phase 3, you will be completing the building of a software system that will enable you to conduct this type of testing.

The result of a *kinship test* will be a likelihood ratio that will essentially be computed using the *likelihood* of the *hypothesis* that says the relationship in question is the case&mdash;the prosecutor's hypothesis, $H_p$ or $H_1$, divided by the *hypothesis* that says the two individuals are *unrelated*&mdash;the defense hypothesis, $H_d$ or $H_2$. So the *likelihood ratio* to be reported shall be computed as $LR = \dfrac{H_p}{H_d}$

To conduct a *kinship test*, you will need to

1. load the *genotype* of the two individuals at each DNA *locus*
1. determine the *relationship type* of the two individuals _(See the Phase 1 writeup)_
1. determine which kinship likelihood ratio (LR) formula to employ for <u>each *locus*</u> based on the *genotype patterns* of the two individuals involved at that *locus*
1. look up the *frequencies* of all alleles used in the corresponding kinship LR formula
 - adjust the frequency of each allele that is <u>not</u> present in a given allele table
1. look up the *kinship coefficients* for the relationship to be tested 

Once you have these components ready, you will be able to compute the LR for the relationship under test.

### Allele frequency adjustment

- You shall adjust the frequency of each allele <u>not</u> present in the provided allele frequency table

- You shall carry out this adjustment on a case-by-case basis and in the order that you encounter each such "unseen" allele
 - The updated allele frequency table for a given locus for a given case shall be discarded for the next kinship test case

- You shall assign a default *allele frequency* value, $\hat{f}$, to each *"unseen"* allele using the formula
 $$
 \hat{f} = \dfrac{5}{N + 5}
 $$
 where $N{=}361$

#### Incorporation of an *"unseen"* allele into an allele frequency table

To incorporate a new frequency value, first, we need to assign a *default* value to the "unseen" allele, $\hat{f}$. In our case, we will use the above-mentioned formula, which evaluates to 0.0136612021857924 for the first "unseen" allele, 10c (See above).

Once we introduce this new allele into our allele frequency table, clearly, the sum of all frequencies (*probabilities*) will no longer be equal to `1.0`. However, we need this sum to be `1.0`! To accomplish this, we must adjust all remaining allele frequency values by multiplying them with an *adjustment factor*, whose formula is provided above. This formula is the direct result of algebra given the need for this adjustment. That is, if the frequency of a new allele is $\hat{f}$, then the remaining ones must add up to $(1 - \hat{f})$. So, we need to scale the remaining frequencies so that their sum gets us this particular value.

Let's prove this adjustment formula.

Suppose we have seven (7) alleles in the original frequency table ($A$, $B$, $C$, $D$, $E$, $F$, $G$) with frequencies $f_A$, $f_B$, and so on.

Clearly,

$$
f_A + f_B + f_C + f_D + f_E + f_F + f_G = 1.0
$$

Now suppose that we encounter an "unseen" allele, H, with a default frequency of $\hat{f}$. This means we will have to adjust each frequency so that we have

$$
\hat{f}_A + \hat{f}_B + \hat{f}_C + \hat{f}_D + \hat{f}_E + \hat{f}_F + \hat{f}_G = (1 - \hat{f})
$$

where the frequencies on the left-hand side of the equation are the *adjusted* versions.

The question is *"How can we have the equation with the original allele frequencies add up to $(1 - \hat{f})$ instead of 1?"*

The solution is simple: Multiply both sides of the equation with $(1 - \hat{f})$!

So, we have:

$$
\Big[f_A + f_B + f_C + f_D + f_E + f_F + f_G\Big] \times (1 - \hat{f}) = 1 \times (1 - \hat{f}) 
$$

So, to carry out the adjustment, we have to multiply each existing allele frequency with the *adjustment factor*, $(1 - \hat{f})$.

The second leftmost table in the example provides the final adjusted values with all the three (3) "unseen" alleles.

Note that only the very last "unseen" allele has a frequency value equal to $(1 - \hat{f})$ since all remaining ones&mdash;including the ones for previously encountered "unseen" alleles&mdash;have also been adjusted as described. So $\hat{f}_A = f_A \times (1 - \hat{f})$ and so on.

The three tables to the right of the thick vertical line show the allele frequency adjustment after incorporating "unseen" allele 10c, 11c, and finally 13c in that order.

Note that, in the last table, the allele frequency value of 13c is equal to $\dfrac{5}{N + 5}$. The allele frequency of 11c has been adjusted by multiplying the previous default value by $(1 - \hat{f})$, and the allele frequency of 13 has been adjusted for the second time using the same procedure. Specifically:

- Frequency of allele **13c**: $\dfrac{5}{N + 5} = \dfrac{5}{361 + 5} = \dfrac{5}{366} = 0.013661202186$ $\quad$ *(default $\hat{f}$)*

- Frequency of allele **11c**: $\dfrac{5}{N + 5} \times \Big[1 - \dfrac{5}{N + 5}\Big] = \dfrac{5}{366} \times \Big[1 - \dfrac{5}{366}\Big] = 0.013474573741$ $\quad$ $\quad$ *(default $\hat{f}$, adjusted once)*

- Frequency of allele **10c**: $\dfrac{5}{N + 5} \times \Big[1 - \dfrac{5}{N + 5}\Big] \times \Big[1 - \dfrac{5}{N + 5}\Big] = \dfrac{5}{366} \times \Big[1 - \dfrac{5}{366}\Big] \times \Big[1 - \dfrac{5}{366}\Big] = 0.013290494864$ $\quad$ *(default $\hat{f}$, adjusted twice)*
